```python
#classify movies based on excellent, good and average 
#using user defined function
def rating(rating):
    if rating>=7.0:
        return "Excellent"
    elif rating>=6.0:
        return "Good"
    else:
        return "Average"
```

```python
data['rating_cat']=data['Rating'].apply(rating)
```
